## Here i performed collabrative filtering here i recomended posts for a particular user 
## based on other similar users own choices

```python
import pandas as pd 
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df_users = pd.read_csv('users.csv')
df_posts = pd.read_csv('posts.csv')
df_views = pd.read_csv('views.csv')
```

```python
df_posts.columns=["post_id","title","category","post_type"]
df_posts=df_posts.dropna(how='any',axis=0) 
df_posts
```

```python
df_users.columns=["user_id","name","gender","academics"]
df_users
```

```python
df_new=df_users.merge(df_views)
```

```python
df_new=df_new.merge(df_posts)
```

```python
df_new
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.metrics.pairwise import linear_kernel, cosine_similarity
```

```python
df_new['combined_features'] = df_new['academics'] + df_new['category'] + df_new['post_type']
```

```python
df_new
```

```python
tf = TfidfVectorizer(analyzer='word',ngram_range=(1, 2),min_df=0, stop_words='english')
```

```python
tfidf_matrix = tf.fit_transform(df_new['combined_features'])
```

```python
tfidf_matrix.shape
```

```python
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
cosine_sim[0]
```

```python
df_new = df_new.reset_index()
userid = df_new['user_id']
indices = pd.Series(df_new.index, index=df_new['user_id'])
```

```python
def get_recommendations_userwise(userid1):
    idx = indices[userid1]
    #print (idx)
    sim_scores = list(enumerate(cosine_sim[idx]))
    #print (sim_scores)
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    user_indices = [i[0] for i in sim_scores]
    #print (user_indices)
    a=user_indices[0:21]
    print(a)
    b=[]
    for i in user_indices:
        b.append(userid[i])
    b=set(b)
    return list(b)
```

```python

print ("-----All Similar users post ID------")
get_recommendations_userwise(1)
```

```python
def get_user_id(usrid_list):
    post_userwise = df_new['user_id'].isin(usrid_list) #
    #print(post_userwise)
    df1 = pd.DataFrame(data = df_new[post_userwise], columns=['user_id'])
    postlist = df1['user_id'].tolist()
    post_list = df_new['user_id'].isin(postlist) 
    df_temp = pd.DataFrame(data = df_new[post_list], columns=['title','post_type'])
    return df_temp.drop_duplicates()
```

```python
print("Recomendation for",df_new["name"].iloc[1])
get_user_id(get_recommendations_userwise(1))
```
